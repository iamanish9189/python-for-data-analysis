```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
batsman.info()
```

```python
# plotting multiple plot
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit sharma vs virat kohli carrer comp')
plt.xlabel("Season")
plt.ylabel("Runs_score")
```

```python
#  colors(hex) and line(width and style) and marker(size)
# hidden parameter 'color'
# hidden parameter 'linestyle'
# hex code on google search
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'solid',linewidth = 3)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 2)
plt.title('Rohit sharma vs virat kohli carrer comp')
plt.xlabel("Season")
plt.ylabel("Runs_score")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'solid',linewidth = 3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 2,marker='o')
plt.title('Rohit sharma vs virat kohli carrer comp')
plt.xlabel("Season")
plt.ylabel("Runs_score")
```

```python
# legend -- location
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'solid',linewidth = 3,marker='D',markersize=10,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color = 'black',linestyle = 'dashdot',linewidth = 2,marker='o',markersize=7,label='Rohit')
plt.title('Rohit sharma vs virat kohli carrer comp')
plt.xlabel("Season")
plt.ylabel("Runs_score")
plt.legend()
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
#  it gives overlap
```

```python
#multiple bar charts
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+ 0.2,df['2017'],width=0.2)
```

```python
# xticks
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+ 0.2,df['2017'],width=0.2)


# xticks --replace number to batsman(name)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# problem
children = [10,20,30,10,40]
colors = ['red red red red red red red','blue blue blue blue blue blue blue','green green green green green ','yellow yellow yellow yellow ','pink pink pink pink']
plt.bar(colors,children,color = 'orange')
# category overlap  ----> isse bachne ke liye using xtick or uska parameter rotation='vertical
plt.xticks(rotation='vertical')
plt.show()
```

```python
# stacked bar charts
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')
plt.legend()
# plt.grid()
plt.show()
```

```python
# dataset
df = pd.read_csv("/content/gayle-175.csv")
df
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','blue','orange','green','pink','yellow'])
plt.show()
```
