```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('/content/iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
# cmap = "winter" -- use to control colors
# alpha -- apke graph ka opacity kitna hoga
plt.xlabel('Sepal lenght')
plt.ylabel('petal lenght')
plt.colorbar()
# plt.colorbar() ---- ye batata hai kaun sa color kaun sa value represent kar raha hai
```

```python
# plot size
# it is used to increase the size of the plot
plt.figure(figsize=(15,7))


plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
plt.xlabel('Sepal lenght')
plt.ylabel('petal lenght')
plt.colorbar()
```

```python
batter.head()
```

```python

plt.figure(figsize=(15,6))
plt.scatter(batter['avg'],batter['strike_rate'],color='green',marker='+')
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Stricke Rate')
plt.show()
```

```python
fig,ax =plt.subplots(figsize=(15,6))
ax.scatter(batter['avg'],batter['strike_rate'],color='green',marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strick Rate')
fig.show()
```

```python
# batter dataset
# 1st way
fig,ax =plt.subplots(nrows=2,ncols=1,figsize=(10,6),sharex=True)
ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('Avg vs strike_Rate')
ax[0].set_ylabel('Strike_Rate')

ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# 2nd  way
fig,ax =plt.subplots(nrows=2,ncols=2,figsize=(10,6))
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0,1].scatter(batter['avg'],batter['runs'])

ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

```python
# 3rd way
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')

ax2 =fig.add_subplot(2,2,2)
ax2.hist(batter['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'])
```

**3D Scatter plot**

```python
fig = plt.figure(figsize=(10,6))

ax = plt.subplot(projection='3d')

ax.scatter(batter['avg'],batter['runs'],batter['strike_rate'])
ax.set_title('IPL Batsman Analysis')
ax.set_xlabel('Avg')
ax.set_ylabel('Runs')
ax.set_zlabel('SR')
```

**3D surface plot**

```python
x= np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 +yy**2
```

```python
fig =  plt.figure(figsize=(10,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

ig =  plt.figure(figsize=(10,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

Contour plot

```python
fig =  plt.figure(figsize=(12,8))

ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
# contour_fill
fig =  plt.figure(figsize=(10,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig =  plt.figure(figsize=(10,8))

ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
```

```python
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.colorbar()
# over number 0-19 to 1 to 20 shift using yticks
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
# ballnumber
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.show()
```
